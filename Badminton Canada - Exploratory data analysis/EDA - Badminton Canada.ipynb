```python
import os
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Removal of extra columns and addition of year, average points scored per tournament, separation of MemberID and 
# Player names for doubles category.
```

```python
lengthoflocalrank = file["Rank"].str.len()
file['lenstr'] = lengthoflocalrank
```

```python
file_up = pd.concat((file['Rank'], file['Player'],file.iloc[:, -5:]), axis=1)
file_up
```

```python
year = [0 for i in range(len(file_up['Rank']))]
```

```python
te = 2014
for i in range(len(file_up['Rank'])):
    year[i] = te
    if i==len(file_up['Rank'])-1:
        #print(i)
        break
    if (file_up['lenstr'][i]>30) and (file_up['Rank'][i+1]=='1'):
        #print(te)
        te+=1
```

```python
file_up['Year'] = year
file_up
```

```python
file_up = file_up[file_up['lenstr']<50]
file_up['Avg per tourn'] = (file_up['Points']).astype(int)//(file_up['Tournaments']).astype(int)

state = [0 for i in range(len(file_up['Rank']))]

for i in range(len(file_up['Rank'])):
    state[i] = file_up['Member ID'].iloc[i][:2]

file_up['State'] = state
file_up
```

```python
pl1 = ['a' for i in range(len(file_up['Rank']))]
pl2 = ['b' for i in range(len(file_up['Rank']))]
for i in range(len(file_up['Player'])):
    #print(i)
    for j in range(len(file_up['Player'].iloc[i])-1):
        #print(j)
        if ord(file_up['Player'].iloc[i][j])>96 and 64<ord(file_up['Player'].iloc[i][j+1])<92:
            pl1[i] = file_up['Player'].iloc[i][:j+1]
            pl2[i] = file_up['Player'].iloc[i][j+1:]
            #print('ckeck')
            break
        if ord(file_up['Player'].iloc[i][j])>96 and 125<ord(file_up['Player'].iloc[i][j+1]):
            pl1[i] = file_up['Player'].iloc[i][:j+1]
            pl2[i] = file_up['Player'].iloc[i][j+1:]
            #print('ckeck')
            break

            
file_up['Player1'] = pl1
file_up['Player2'] = pl2
file_up
```

```python
pl1id = [ 'mid' for i in range(len(file_up['Member ID']))]
pl2id = [ 'mid' for i in range(len(file_up['Member ID']))]
for i in range(len(file_up['Member ID'])):
    #print(i)
    n = len(file_up['Member ID'].iloc[i])
    for j in range(n-1):
    #print(j)
        if ord(file_up['Member ID'].iloc[i][j])<60 and 60<ord(file_up['Member ID'].iloc[i][j+1]):
            pl1id[i] = file_up['Member ID'].iloc[i][:j+1]
            pl2id[i] = file_up['Member ID'].iloc[i][j+1:]
            #print('ckeck')
            break
```

```python
file_up['Player 1 ID'] = pl1id
file_up['Player 2 ID'] = pl2id
```
